```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as clr
from sklearn.model_selection import train_test_split
# generate random floating point values
from random import seed
from random import random
```

```python
def f(x):
    return -x+5

def generateData(n):
    seed(1)
    min=0
    max=5
    L=[]
    for _ in range(n):
        valueX= random()
        scaledValueX=min + (valueX * (max - min))
        x= round(scaledValueX,1)
        
        valueY= random()
        scaledValueY=min + (valueY * (max - min))
        y= round(scaledValueY,1)
        
        y0=f(x)
        cl= 1 if y>y0 else 0
        L.append([x,y,cl])
    return L
```

```python
#generate data
data = np.array(generateData(100))
# split data
X1 = data[:,0]
X2 = data[:,1]
C = data[:,2]
colors = ['red','blue']
# plot graph
plt.scatter(X1,X2,c=C,cmap=clr.ListedColormap(colors))
plt.show()
```

```python
X = [[X1[k],X2[k]] for k in range(len(C))]
# splitting data for learning and testing
X_train,X_test,y_train,y_test = train_test_split(X, C, test_size=0.33, random_state=42)
X_train=np.array(X_train)
X_test=np.array(X_test)
x1_train=X_train[:,0]
x2_train=X_train[:,1]
```

```python
def get_indexes(C,predicted_class):
    indexes=[]
    for i in range(len(C)):
        if(C[i]!=predicted_class[i]):
            indexes.append(i)
    return indexes
```

```python
def getX2(x1,bias,w1,w2):
    return -(w1/w2)*x1-bias/w2


def plot_data_line(X1,X2,C,predicted_class,bias,w1,w2):
    # plotting the points
    plt.figure(figsize=(5,3))
    colors= ['red','blue']
    plt.scatter(X1,X2,c=C,cmap=clr.ListedColormap(colors))
    
    # get indexes where is a wrong prediction
    error_indexes= get_indexes(C,predicted_class)
    
    #plotting the wrongful points in black color
    plt.plot(X1[error_indexes],X2[error_indexes],'ko')
    
    # draw the seperable line
    x1_A=0
    x2_A= getX2(x1_A,bias,w1,w2)
    x1_B=5
    x2_B= getX2(x1_B,bias,w1,w2)
    plt.plot([x1_A,x2_A],[x1_B,x2_B],color="green")
    
    plt.show()
```

```python
# fitting data
def segmoid(Z):
    return 1/(1+np.exp(-Z))

def fit(X1,X2,y):
    lr=0.01
    n=len(y)
    epochs=10001
    w1=0.5
    w2=0.5
    bias=0.5
    
    for i in range(epochs):
        ws= bias + w1*X1 + w2*X2
        yhat= segmoid(ws)
        predicted_class= np.array([1 if yh>0.5 else 0 for yh in yhat])
        if(i%100==0):
            print("iter: ",i)
            plot_data_line(X1,X2,y,predicted_class,bias,w1,w2)
        
        dbias=-(2/n)*np.sum(y-yhat)
        dw1=-(2/n)*np.sum(X1*(y-yhat))
        dw2=-(2/n)*np.sum(X2*(y-yhat))
        
        bias -= dbias*lr
        w1 -= dw1*lr
        w2 -= dw2*lr
    return bias,w1,w2
    
```

```python
bias,w1,w2 = fit(x1_train,x2_train,y_train)
```

```python
def predict(X1,X2):
    ws = bias+w1*X1+w2*X2
    y_predicted = segmoid(ws)
    return np.array([1 if yp>0.5 else 0 for yp in y_predicted])
```

```python
predicted_class= predict(X_test[:,0], X_test[:,1])
for pr,accurate in zip(predicted_class, y_test):
    if pr!=accurate:
        print(pr," ",accurate)

plot_data_line(X_test[:,0], X_test[:,1], y_test, predicted_class, bias, w1, w2)
```
